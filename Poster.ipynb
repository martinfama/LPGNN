```python
import sys
import importlib
import imageio
from tqdm import tqdm

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import igraph as ig
import networkx as nx
import torch as th
import torch_geometric as pyg

import LPGNN
```

```python
# using networkX, generate a random of 6 nodes, and draw it using labels for the nodes
G = nx.gnp_random_graph(6, 0.5)
# enlarge the nodes and the label size
fig, ax = plt.subplots(figsize=(10, 10))
nx.draw(G, with_labels=True, node_size=5000, font_size=50, ax=ax)
# save figure as png, with high resolution, at /figs/Poster/
plt.savefig('figs/Poster/graph.png', dpi=300)
```

```python
# generate a Barabasi-Albert graph with 100 nodes, and draw it without labels
G = nx.barabasi_albert_graph(50, m=3)
fig, ax = plt.subplots(figsize=(10, 10))
# set node size to the degree of the node
nx.draw(G, with_labels=False, node_size=[G.degree(n) * 20 for n in G.nodes], ax=ax)
plt.savefig('figs/Poster/graph_BA.png', dpi=300)
```
